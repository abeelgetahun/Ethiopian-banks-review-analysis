```python
# Libraries for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Jupyter settings
%matplotlib inline
sns.set(style="whitegrid")
```

```python
import os

print("Files in data/processed folder:", os.listdir("data/processed"))
```

```python
data_dir = "data/processed"

boa_df = pd.read_csv(f"{data_dir}/boa_reviews_sentiment.csv")
cbe_df = pd.read_csv(f"{data_dir}/cbe_reviews_sentiment.csv")
dashen_df = pd.read_csv(f"{data_dir}/dashen_reviews_sentiment.csv")

# Just to confirm they loaded successfully, print their shapes:
print("BOA DataFrame shape:", boa_df.shape)
print("CBE DataFrame shape:", cbe_df.shape)
print("Dashen DataFrame shape:", dashen_df.shape)
```

```python
# Add a 'bank' column to each dataframe
boa_df['bank'] = 'BOA'
cbe_df['bank'] = 'CBE'
dashen_df['bank'] = 'Dashen'

# Combine the three dataframes
all_reviews = pd.concat([boa_df, cbe_df, dashen_df], ignore_index=True)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=all_reviews, x='bank', y='sentiment')
plt.title("Sentiment Score Distribution per Bank", fontsize=14)
plt.xlabel("Bank")
plt.ylabel("Sentiment Score")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=all_reviews, x="rating", hue="bank", bins=5, multiple="dodge", palette="muted")
plt.title("Rating Distribution per Bank", fontsize=14)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=all_reviews, x='bank', y='review_length', palette='Pastel1')
plt.title(" Distribution of Review Lengths by Bank", fontsize=14)
plt.xlabel("Bank")
plt.ylabel("Review Length (characters or words)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=all_reviews, x='sentiment', hue='bank', multiple='stack', palette='Set1', bins=30)
plt.title("Distribution of Sentiment Scores by Bank", fontsize=14)
plt.xlabel("Sentiment Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=all_reviews, x='review_length', y='sentiment', hue='bank', palette='Dark2', alpha=0.7)
plt.title("Sentiment vs. Review Length by Bank", fontsize=14)
plt.xlabel("Review Length")
plt.ylabel("Sentiment Score")
plt.show()
```
